```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy as sp
```

```python
csn = pd.read_excel("uitslag_csn.xlsx")
```

# 1
Bepaal de gemiddelde scores per student en toon de hoogstscorende eerst.

```python
csn_results = pd.DataFrame([v.mean()*10 for v in csn.values[:, [csn.columns.get_loc(v) for v in csn.columns if v.startswith("Score")]]], columns=["Mean"])
csn_results.sort_values(by="Mean", ascending=False)
```

# 2
Voeg twee kolommen per student toe
1. cijfer = $\frac{9}{30}(\sum_{i=1}^{40}{Score.i}-10) + 1$
2. eindcijfer = $\max(1, cijfer)$. (Tip: gebruik ***np.maximum***)

```python
csn_results["Cijfer"] = [9/30*(v.mean()*10-10)+1 for v in csn.values[:, [csn.columns.get_loc(v) for v in csn.columns if v.startswith("Score")]]]
#csn_results["Eindcijfer"] = [max(1, 9/30*(v.mean()*10-10))+1 for v in csn.values[:, [csn.columns.get_loc(v) for v in csn.columns if v.startswith("Score")]]]
csn_results["Eindcijfer"] = np.maximum(1, csn_results["Cijfer"])
csn_results
```

# 3
Bepaal het totale gemiddelde over alle studenten. Doe dit nogmaals, maar dan alleen de studenten die een voldoende hebben gescoord.

```python
csn_win = pd.DataFrame([True if v.mean()*10 >= 5.5 else False for v in csn.values[:, [csn.columns.get_loc(v) for v in csn.columns if v.startswith("Score")]]], columns=["Mean"])
csn_win
```

```python
mu = csn_results.Eindcijfer.mean()
sigma = csn_results.Eindcijfer.std()
normal_verdeling = sp.stats.norm(mu, sigma)

x = np.linspace(-5, 15, 1000)
y = normal_verdeling.pdf(x)

plt.plot(x, y)
plt.show()
```
