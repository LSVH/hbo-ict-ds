# 2.3
Van welk type is iedere dimensie (nominaal/categoriaal/discreet/continue)? Plaats de antwoorden in een tabel.

```python
csn_cols = csn.columns
csn_dtype = OrderedDict({"Column":list(csn_cols), "Nominal":[], "Categorical":[], "Discrete":[], "Continuous":[]})
csn_categorical = list(set(csn_cols)-set(csn._get_numeric_data().columns))
for col in range(csn.shape[1]):
    csn_dtype["Nominal"].append(list(set(isinstance(s, str) for s in csn.values[:, col]))[0])
    csn_dtype["Categorical"].append(True if csn_cols[col] in csn_categorical else False)
    csn_dtype["Discrete"].append(True if len(csn.values[:, col]) == len(list(set(csn.values[:, col]))) else False)
    csn_dtype["Continuous"].append(True if csn.values[:, col].min() < csn.values[:, col].max() else False)

pd.DataFrame.from_dict(csn_dtype)
```

```python
prog_cols = prog.columns
prog_dtype = OrderedDict({"Column":list(prog_cols), "Nominal":[], "Categorical":[], "Discrete":[], "Continuous":[]})
prog_categorical = list(set(prog_cols)-set(prog._get_numeric_data().columns))
for col in range(prog.shape[1]):
    prog_dtype["Nominal"].append(list(set(isinstance(s, str) for s in prog.values[:, col]))[0])
    prog_dtype["Categorical"].append(True if prog_cols[col] in prog_categorical else False)
    prog_dtype["Discrete"].append(True if len(prog.values[:, col]) == len(list(set(prog.values[:, col]))) else False)
    prog_dtype["Continuous"].append(True if prog.values[:, col].min() < prog.values[:, col].max() else False)

pd.DataFrame.from_dict(prog_dtype)
```

# 2.4
De excel-file bevatte dezelfde header-name voor alle scores. De importer heeft dit voor je opgelost door duplicate header-namen te nummeren. Nadeel is echter dat de eerste zelf geen nummer heeft. Manipuleer de headers zodat de telling begint met Score.1 ipv. Score (merk op dat Score.1 dan hernoemt moet worden naar Score.2, etc.).

```python
csn_scores = [a for a in csn.columns if a.startswith("Score")]
csn_non_scores = [ns for ns in csn.columns if ns not in csn_scores]
csn_scores = ["Score."+str(b+1) for b in range(0, len(csn_scores))]
csn.columns = csn_non_scores+csn_scores
csn.columns
```

```python
prog_scores = [a for a in prog.columns if a.startswith("Score")]
prog_non_scores = [ns for ns in prog.columns if ns not in prog_scores]
prog_scores = ["Score."+str(b+1) for b in range(0, len(prog_scores))]
prog.columns = prog_non_scores+prog_scores
prog.columns
```
